## Define the Problem ##

We want to predict the per capita income of each county in Arkansas based upon the educational attainment of that county.  

We have data that tells the educational attainment of each county, but we don't know (at least in this example) the per capita income for each county.

We are going to build a model to predict per capita income of a population based upon the educational attainment of that population.

More specifically, we are going to use data we have that maps per capita income of each state to the educational attainment in that state.

We are hypothesizing that we can build a good linear regression model using this state level data to build a predictor that given the educational attainment as input
will predict the per capita income of that population.  Furthermore, we are hypothesizing that a model built upon state level data is a good predictor for counties in Arkansas.

### Make a useful helper function ###

We are going to end up trying a few different options to build our model.  Let's use the power of Python functions to make a handy routine build and visualize a model.

This function uses the Python ***skilearn*** module to build our model and the ***matplotlib*** module for graphing.  These are excellent examples of how existing Python modules (i.e., collections of reusable library functions) greatly increase our productivity in writing code.

```python
def build_linear_regression_model(x, y, 
                                  xlabel='X (Independent Variable)', ylabel='Y (Dependent Variable)', title=None, 
                                  figure_size = [10, 6], data_color='b', prediction_color='r', fontsize=14):
    """
    Build a linear regression model from the input x and y.  Provide a visualization of the resulting data and model.
    
    Parameters:
            x:                list of x values to use for training
            y:                list of y values to use for training (elements correspond to elements in x)
            xlabel:           label for x-axis (defaulted)
            ylabel:           label for y-axis (defaulted)
            title:            chart title (optional)
            figure_size:      list containing horizontal and vertical size (e.g., [10, 6]) for figure (defaulted)
            data_color:       color for (x,y) scatter graph of training data points (defaulted)
            prediction_color: color for regression line (predictions) and regression equation (defaulted)
            fontsize:         text font size (defaulted)
            
    Returns:
            model:            a trained linear regression model
    """
    
    plt.figure(figsize=tuple(figure_size))

    # convert x and y to a 2D numpy array (required for input to the 'fit' method below)
    X_train = np.c_[x]
    y_train = np.c_[y]
    
    # Show the training data on a scatter plot
    plt.scatter(X_train, y_train, color=data_color)

    # create a linear regression model object
    model = sklearn.linear_model.LinearRegression()
    
    # train our model using our training data; this figures out the best regression line for the training data
    model.fit(X_train, y_train)
    
    # extract the slope and intercept from the model for the best regression line
    intercept, slope = model.intercept_[0], model.coef_[0][0]
    
    # compute the r squared value for the model against the training data
    r_squared = model.score(X_train, y_train)
    
    # plot the training X data using the regression line
    plt.plot(X_train, intercept + slope * X_train, prediction_color)
    
    # compute a good spot to place the line equation and r squared value on the plot
    textpos = [min(X_train) + (max(X_train) - min(X_train)) * 0.33, max(y_train)]
    line_text = f'y = {intercept:.4f} + {slope:.4f} * x'
    plt.text(textpos[0], textpos[1], line_text, fontsize=fontsize, color=prediction_color)
    line_text = f'$r^2 = {r_squared:.4f}$'
    plt.text(textpos[0], textpos[1] - ((max(y_train) - min(y_train)) * 0.1), line_text, fontsize=fontsize, color=prediction_color)

    if title is not None:
        plt.title(title)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
    
    return model
```

Defining the "docstring" portion of the function above (the multi-line comment after the function definition above) is good Python programming practice.  It documents the function and enables us to invoke ***help*** on the function.

```python
help(build_linear_regression_model)
```

Demonstrate our function using a perfect regression situation (i.e., all the training data points form a line).

```python
sample_model = build_linear_regression_model(x=[1, 2, 3, 4, 5], y=[1, 2, 3, 4, 5], title='Sample Regression')
```

We have three levels of education attainment:

Let's explore the options a bit using our modeling function to see what seems the best path forward.

### Possible models for Household Income ###

```python
possible_x = ['High school', 'Bachelors', 'Advanced']
y = 'HouseholdIncome'
```

```python
for x in possible_x:
    model = build_linear_regression_model(x=education_income[x], y=education_income[y], xlabel=x, ylabel=y)
```

### Possible models for Per Capita Income ###

```python
possible_x = ['High school', 'Bachelors', 'Advanced']
y = 'PerCapitaIncome'
```

```python
for x in possible_x:
    model = build_linear_regression_model(x=education_income[x], y=education_income[y], xlabel=x, ylabel=y)
```

Our best model uses "Bachelors" to predict "PerCapitaIncome".

Set our base model to this combination.

```python
x = 'Bachelors'
y = 'PerCapitaIncome'
model = build_linear_regression_model(x=education_income[x], y=education_income[y], xlabel=x, ylabel=y)
```
